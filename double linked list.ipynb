```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.previous =None
class doubleLL:
    def __init__(self):
        self.head = None
    def printLL(self):
        if self.head is None:
            print("Linked List is Empty")
        else:
            lastNode = self.head
            while lastNode is not None:
                print(lastNode.data)
                lastNode = lastNode.next
                
    def printLLrev(self):
        if self.head is None:
            print("Its Empty")
        else:
            currentNode = self.head
            while currentNode.next is not None:
                currentNode = currentNode.next
            while currentNode is not None:
                print(currentNode.data)
                currentNode = currentNode.previous
                
    def insert_empty(self,data):
        if self.head is None:
            newNode = Node(data)
            self.head = newNode
        else:
            print("Linkedlist is not Empty")
        
        
        
    def add_began(self,data):
        newNode = Node(data)
        if self.head is None:
            self.head = newNode
        else:
            newNode.next = self.head
            self.head.previous = newNode
            self.head = newNode
            
    def add_end(self,data):
        newNode = Node(data)
        if self.head is None:
            self.head = newNode
        else:
            currentNode = self.head
            while currentNode is not None:
                currentNode = currentNode.next
            currentNode.next = newNode
            newNode.previous = currentNode
            
dll = doubleLL()
dll.insert_empty(10)

dll.add_began(20)
dll.add_end(30)

dll.printLL()
print("Linkedlist reverse is:")
dll.printLLrev()
```
